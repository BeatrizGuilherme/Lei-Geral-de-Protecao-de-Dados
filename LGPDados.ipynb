```python
# Importando as bibliotecas
import pandas as pd
import hashlib
```

```python
# 1. Criar dataset fictício com dados pessoais
df = pd.DataFrame({
    "nome": ["Ana Silva", "Marcos Souza", "Julia Mendes", "Carlos Rocha"],
    "email": ["ana@gmail.com", "marcos@hotmail.com", "julia@outlook.com", "carlos@gmail.com"],
    "telefone": ["11987654321", "21999995555", "11912344321", "21988887777"],
    "cpf": ["12345678901", "98765432100", "45678912300", "32165498700"],
    "idade": [25, 34, 29, 42],
    "salario": [3500, 7200, 4800, 9100]
})

print("**BASE ORIGINAL**")
print(df)
```

```python
# 2. Função para anonimizar (hash seguro SHA-256)
def hash_value(value):
    return hashlib.sha256(value.encode()).hexdigest()
```

```python
# 3. Aplicar anonimização
df_anon = df.copy()

# Hash de identificadores
df_anon["id_hash"] = df["cpf"].apply(hash_value)
```

```python
# Mascarar e-mails (ex: a***@gmail.com)
df_anon["email"] = df["email"].apply(lambda x: x[0] + "***@" + x.split("@")[1])
```

```python
# Mascarar telefone (ex: ******4321)
df_anon["telefone"] = df["telefone"].apply(lambda x: "*" * 6 + x[-4:])
```

```python
# Remover dados extremamente sensíveis (LGPD recomenda minimização)
df_anon = df_anon.drop(columns=["nome", "cpf"])

print("\n**BASE ANONIMIZADA**")
print(df_anon)
```
